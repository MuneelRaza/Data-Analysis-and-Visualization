Seaborn is a **Python data visualization library based on matplotlib**. It provides a high-level interface for drawing attractive and informative statistical graphics. The library provides a database of useful datasets for educational purposes that can be loaded by typing:

`seaborn.load_dataset(name)`

For information on what these datasets are : https://github.com/mwaskom/seaborn-data. 

We will be working with a famous titanic data set for these exercises.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('whitegrid')
titanic = sns.load_dataset('titanic')
titanic.head()
```

# **Exercises**
Recreate the plots below using the titanic dataframe. Also explain which kind of plot is used and what relationship between the mapped data it is explaining. Pay special attention to x and y labels of plot to get idea of what kind of data to plot.

**TASK 1:**

LAB5_Task1.PNG

```python
#code start here
sns.histplot(titanic["fare"], bins = 30, color='red')
#code ends here
```

The graph above is a histogram that is used to show frequency of intervals. The histogram is a univariate plot type which accepts only one column and their is a frequency column on y-axis. The above graph shows the frequency of fares. The fare column is divided into suitable intervals by the function histplot and on y-axis the frequency of respective interval is shown.

**TASK 2:**

LAB5_Task2.PNG

```python
#code starts here
sns.boxplot(x="class", y="age", data=titanic)

# Set plot title and axes labels
plt.xlabel("class")
plt.ylabel("age")
plt.show()
#code ends here
```

A boxplot shows the summary of set of data using five numbers these are minimum, first quartile, median, third quartile, and maximum. Here, in the first class the minimum is at almost median at around 38 and maximum at around 80. Similarly, in the second and third classes we have same five numbers but the marks below and above shows outliers.

```python
#code starts here
sns.swarmplot(x=titanic["class"], y=titanic["age"])
#code ends here
```

Swarmplot shows the numeric distributions of data over categories which is very similar to boxplot.Here, we have three categories, the dense part shows medain area, lower dot shows minimum value and uper dot shows max value.

**TASK 5:**

LAB5_Task5.PNG

```python
#code starts here
cols = ['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare', 'adult_male', 'alone']

# Create a correlation matrix
corr = titanic[cols].corr()

# Create heatmap
sns.heatmap(data=corr, cmap="YlGnBu", linewidths=.5)
#code ends here
```

Heatmap provides visual summary of data using different colors. It allows us to identify critical part in the dataset

```python
#code starts here
sns.jointplot(data=titanic, x='fare', y='age')
#code ends here
```

Jointplot shows the relationship between two variables e:g age and fare in our case along with the distribution of individual varaible.

**TASK 7:**

LAB5_Task7.PNG

```python
#code starts here
graph = sns.FacetGrid(titanic, col='alive', row='alone', height=4)
graph.map(sns.histplot, 'age')
graph.set(xlim=(0, 70), ylim=(0, 80))
plt.show()
#code ends here
```

This graph shows the summary of relationship between different variables using subplots.
